```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

model_name = 'Facenet'

img_base_path = '/content/gdrive/MyDrive/YaleDataBaseB/yaleB11/yaleB11_P00A+000E+00.pgm'
img_ver_path = '/content/gdrive/MyDrive/YaleDataBaseB/yaleB11/yaleB11_P00A-035E-20.pgm' 

img_base = cv2.imread(img_base_path)
img_ver = cv2.imread(img_ver_path)

resp = DeepFace.verify(img_base_path, img_ver_path, model_name= model_name)

resp['distance']
```

```python
figure(figsize=(15, 8), dpi=80)

plt.subplot(1, 2, 1)
plt.imshow(img_base[:, :, ::-1])

plt.subplot(1, 2, 2)
plt.imshow(img_ver[:, :, ::-1])

plt.show()
```
